```python
import pandas as pd
from neuralprophet import NeuralProphet
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("tokyo_weather.csv")

# Prepare data
data = df[["Date", "Highest Temperature (°C)"]].copy()

data["Date"] = pd.to_datetime(data["Date"], errors="coerce")
data["Highest Temperature (°C)"] = pd.to_numeric(
    data["Highest Temperature (°C)"],
    errors="coerce"
)

data = data.rename(columns={
    "Date": "ds",
    "Highest Temperature (°C)": "y"
})

data = data.dropna()
data = data.sort_values("ds")
data = data.drop_duplicates(subset="ds")
data = data.reset_index(drop=True)

print(data.head())
print(data.dtypes)

# IMPORTANT: set learning_rate manually
model = NeuralProphet(
    learning_rate=0.01,
    yearly_seasonality=True,
    weekly_seasonality=True,
    daily_seasonality=False
)

metrics = model.fit(data, freq="D")

print(metrics.tail())
```

```python
from sklearn.metrics import mean_absolute_percentage_error

future = model.make_future_dataframe(
    data,
    periods=15,
    n_historic_predictions=True
)

forecast = model.predict(future)

forecast[["ds", "y", "yhat1"]].tail(30)

mape = mean_absolute_percentage_error(
    comparison_actuals["Actual"],
    comparison_actuals["Predicted"]
) * 100

print("MAPE:", mape, "%")
```
